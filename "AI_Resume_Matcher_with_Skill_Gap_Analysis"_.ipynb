```python

# AI Resume Analyzer (with Matplotlib Visuals)
# Run this in a separate cell before running the code:
# !pip install pypdf python-docx matplotlib scikit-learn
# Install required libraries
!pip install pypdf python-docx matplotlib scikit-learn
import pypdf
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import docx  # python-docx

# 1. Extract Resume Text
def extract_text_from_pdf(file_path):
    """Extract text from a PDF file using pypdf"""
    text = ""
    with open(file_path, "rb") as f:
        reader = pypdf.PdfReader(f)
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
               # check if page is not empty
                text += page_text + " "
    return text.strip()

def extract_text_from_docx(file_path):
    """Extract text from a DOCX file using python-docx"""
    doc = docx.Document(file_path)
    return " ".join([para.text for para in doc.paragraphs])

#  2. Example Resume & Job Description (Demo Data)
resume_text = """
Amit Pawar
Email: pawaramit202004@gmail.com
Skills: Python, Machine Learning, NLP, TensorFlow, SQL
Education: BTech AI/ML at SIT Pune
Experience: Intern at SSVL Technology
"""

job_description = """
We are hiring an AI/ML Engineer with skills in:
- Python, Machine Learning, NLP
- Deep Learning (TensorFlow or PyTorch)
- SQL and Data Analysis
- Good communication skills
"""

#  3. Preprocess
documents = [resume_text, job_description]
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

#  4. Cosine Similarity (Match Score)
similarity_score = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
match_percentage = round(similarity_score * 100, 2)

print(" Resume vs Job Match Score:", match_percentage, "%")

#  5. Matplotlib Visuals
labels = ["Match %", "Gap %"]
sizes = [match_percentage, 100 - match_percentage]
colors = ["#4CAF50", "#FF5252"]

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

#  Bar Chart
ax[0].bar(["Resume Match"], [match_percentage], color="#4CAF50")
ax[0].set_ylim(0, 100)
ax[0].set_ylabel("Percentage")
ax[0].set_title("Resume Match Score")

#  Pie Chart
ax[1].pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, startangle=90)
ax[1].set_title("Match vs Missing Skills")

plt.suptitle("AI Resume Analyzer Results", fontsize=14, fontweight="bold")
plt.show()
```

```python
# AI Resume Analyzer (with Matplotlib Visuals)
# Run this in a separate cell before running the code:
# !pip install pypdf python-docx matplotlib scikit-learn
# Install required libraries
import matplotlib.pyplot as plt
#scikit-learn library, used for converting text data into numerical
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

#1. Input Resume & Job Description
print("Enter Resume Text (press Enter when done):")
resume_text = ""
while True:
    line = input()
    if line.strip() == "":
       # empty line to finish
        break
    resume_text += line + " "

job_description = """
We are hiring an AI/ML Engineer with skills in:
- Python, Machine Learning, NLP
- Deep Learning (TensorFlow or PyTorch)
- SQL and Data Analysis
- Good communication skills
"""

documents = [resume_text, job_description]
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

similarity_score = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
match_percentage = round(similarity_score * 100, 2)

print(f"\n Resume vs Job Match Score: {match_percentage}%")

# 3. Matplotlib Visuals libraey using
labels = ["Match %", "Gap %"]
sizes = [match_percentage, 100 - match_percentage]
colors = ["yellow", "red"] # Corrected color names

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Bar Chart
ax[0].bar(["Resume Match"], [match_percentage], color="#4CAF50")
ax[0].set_ylim(0, 100)
ax[0].set_ylabel("Percentage")
ax[0].set_title("Resume Match Score")

# Pie Chart
ax[1].pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, startangle=90)
ax[1].set_title("Match vs Missing Skills")

plt.suptitle("AI Resume Analyzer Results", fontsize=14, fontweight="bold")
plt.show()
```
